## 1. Average Salary Distribution for Top 10 Paying Data Analyst in 2023

```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_json(r'project_sql\data\first_query.json')
df
```

```python
sns.barplot(df, y='job_title', x='salary_year_avg')
plt.xlabel('Average Yearly Salary ($USD)')
plt.ylabel(' ')
plt.yticks(fontsize=8)
plt.title('Average Salary Distribution for Top 10 Paying Data Analyst in 2023')
```

## Skill Count Top Paying Data Analyst Jobs in 2023

```python
df1 = pd.read_json(r'project_sql\data\second.json')
df1
```

```python
skills = df1['skills'].value_counts()
y = skills.index[:10]
x = skills.values[:10]
```

```python
sns.barplot(df1, y=y, x=x)
plt.xlabel('Frequency')
plt.ylabel(' ')
plt.yticks(fontsize=8)
plt.title('Skill Count Top Paying Data Analyst Jobs in 2023')
```
